## 1. Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

import warnings
warnings.filterwarnings("ignore")
```

## 4. Modeling and Prediction

```python
y = df["Class"]
X = df.drop("Class", axis = 1)
```

### We apply oversampling to the dataset, so we will have balanced data

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)
```

```python
sns.countplot(y_res)
```

#### Our data is now balanced

### We split the data as train and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_res,y_res,test_size = 0.2, random_state = 42)
```

### K Neighbors Classifier

```python
KNC = KNeighborsClassifier()
KNC.fit(x_train, y_train)
pred_KNC = KNC.predict(x_test)
```

```python
print(classification_report(pred_KNC, y_test))
```

```python
sns.heatmap(confusion_matrix(pred_KNC, y_test), annot = True)
```

### Decision Tree Classifier

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
pred_DTC = DTC.predict(x_test)
```

```python
print(classification_report(pred_DTC, y_test))
```

```python
sns.heatmap(confusion_matrix(pred_DTC, y_test), annot = True)
```

### Random Forest Classifier

```python
RFC = RandomForestClassifier()
RFC.fit(x_train, y_train)
pred_RFC = RFC.predict(x_test)
```

```python
print(classification_report(pred_RFC, y_test))
```

```python
sns.heatmap(confusion_matrix(pred_RFC, y_test), annot = True)
```

### We can also apply undersampling to the dataset, so we will again have balanced data

```python
from imblearn.under_sampling import RandomUnderSampler

ros = RandomUnderSampler()
X_res, y_res = ros.fit_resample(X, y)
```

```python
sns.countplot(y_res)
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(X_res,y_res,test_size = 0.2, random_state = 42)
```

### K Neighbors Classifier

```python
KNC = KNeighborsClassifier()
KNC.fit(x_train, y_train)
pred_KNC = KNC.predict(x_test)
```

```python
print(classification_report(pred_KNC, y_test))
```

```python
sns.heatmap(confusion_matrix(pred_KNC, y_test), annot = True)
```

### Decision Tree Classifier

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
pred_DTC = DTC.predict(x_test)
```

```python
print(classification_report(pred_DTC, y_test))
```

```python
sns.heatmap(confusion_matrix(pred_DTC, y_test), annot = True)
```

### Random Forest Classifier

```python
RFC = RandomForestClassifier()
RFC.fit(x_train, y_train)
pred_RFC = RFC.predict(x_test)
```

```python
print(classification_report(pred_RFC, y_test))
```

```python
sns.heatmap(confusion_matrix(pred_RFC, y_test), annot = True)
```

#### We achieved 99% accuracy when we oversampled the data, and we achieved 94% accuracy when we undersampled the data
